```python
import pandas as pd
```

```python
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_test = pd.read_csv("y_test.csv")
y_train = pd.read_csv("y_train.csv")

X_train.drop(["Unnamed: 0"],axis =1,inplace = True)
X_test.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# TF-IDF vektörleştirme
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(y_train['pl_name'])

# Kosinüs benzerliği hesaplama
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Her bir pl_name için benzerlik skorlarına göre bir sayı ata
similarity_threshold = 0.9  # Belirli bir benzerlik eşik değeri
assigned_numbers = {}
number_counter = 1

for i in range(len(y_train)):
    if i not in assigned_numbers:
        assigned_numbers[i] = number_counter
        number_counter += 1
    for j in range(i+1, len(y_train)):
        if cosine_sim[i][j] >= similarity_threshold:
            assigned_numbers[j] = assigned_numbers[i]

# Yeni bir "assigned_number" sütunu oluşturun ve sonuçları ekleyin
y_train['assigned_number'] = y_train.index.map(assigned_numbers)

# Sonuçları göster
print(y_train)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Örnek veri çerçevesi
data = {'pl_name': pro['pl_name']}
df = pd.DataFrame(data)

# TF-IDF vektörleştirme
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['pl_name'])

# Kosinüs benzerliği hesaplama
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Her bir pl_name için benzerlik skorlarına göre bir sayı ata
similarity_threshold = 0.3  # Belirli bir benzerlik eşik değeri
assigned_numbers = {}
number_counter = 1

for i in range(len(df)):
    if i not in assigned_numbers:
        assigned_numbers[i] = number_counter
        number_counter += 1
    for j in range(i+1, len(df)):
        if cosine_sim[i][j] >= similarity_threshold:
            assigned_numbers[j] = assigned_numbers[i]

# Sonuçları yeni bir sütun olarak ekleyin
df['assigned_number'] = df.index.map(assigned_numbers)

# Sonuçları göster
print(df)
```
